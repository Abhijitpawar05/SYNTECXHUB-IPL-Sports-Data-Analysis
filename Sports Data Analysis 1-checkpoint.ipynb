```python
#Import required Libelary
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
import pandas as pd

matches = pd.read_csv(r"C:\Users\USER\Downloads\matches (1).csv")
deliveries = pd.read_csv(r"C:\Users\USER\Downloads\deliveries (2).csv")

print(matches.shape)
print(deliveries.shape)

matches.head()
```

```python
#Step 4:- Top 10 Run Scorers Analysis
# Grouped data by batsman.
# Calculated total runs scored by each player.
# Sorted in descending order.
# Visualized using bar chart.
# Insight: Identified highest run scorers in IPL history.
top_scorers = deliveries.groupby("batsman")["batsman_runs"].sum().sort_values(ascending=False).head(10)

print(top_scorers)
```

```python
#In Visulization
top_scorers.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Run Scorers in IPL")
plt.ylabel("Total Runs")
plt.show()
```

```python
### Step 5: Strike Rate Calculation
# Counted total balls faced by each batsman.
# Calculated strike rate using formula:
# Strike Rate = (Total Runs / Total Balls) * 100
# Sorted top strike rate players.
# Visualized using bar plot.
# Insight: Identified aggressive batsmen.
balls = deliveries.groupby("batsman")["ball"].count()
runs = deliveries.groupby("batsman")["batsman_runs"].sum()

strike_rate = (runs / balls) * 100
strike_rate = strike_rate.sort_values(ascending=False).head(10)

print(strike_rate)
```

```python
#Plot
strike_rate.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Strike Rates")
plt.ylabel("Strike Rate")
plt.show()
```

```python
# Step 6: Team Win Percentage Analysis
# Counted total matches won by each team.
# Counted total matches played by each team.
# Calculated Win % using:
# Win % = (Matches Won / Matches Played) * 100
# Visualized team performance.
# Insight: Identified most successful IPL teams
total_matches = matches["winner"].value_counts()
total_played = pd.concat([matches["team1"], matches["team2"]]).value_counts()

win_rate = (total_matches / total_played) * 100
win_rate = win_rate.sort_values(ascending=False)

print(win_rate)
```

```python
#Plot
win_rate.plot(kind='bar', figsize=(12,6))
plt.title("Team Win Percentage")
plt.ylabel("Win %")
plt.show()
```

```python
# Step 7: Season-wise Run Trend Analysis
# Merged deliveries and matches dataset on match_id.
# Calculated total runs scored per season.
# Visualized using line chart.
# Insight: Observed scoring trend changes across seasons.
season_runs = deliveries.merge(matches[['id','season']], left_on='match_id', right_on='id')

season_total = season_runs.groupby("season")["total_runs"].sum()

season_total.plot(kind='line', marker='o', figsize=(10,5))
plt.title("Total Runs Per Season")
plt.ylabel("Total Runs")
plt.show()
```

```python
# Step 8: Player Performance Across Seasons
# Filtered specific player data (e.g., Virat Kohli).
# Calculated season-wise total runs.
# Plotted line graph.
# Insight: Compared player consistency across seasons
kohli = season_runs[season_runs["batsman"] == "V Kohli"]

kohli_season = kohli.groupby("season")["batsman_runs"].sum()

kohli_season.plot(kind='line', marker='o', figsize=(8,4))
plt.title("Virat Kohli Season-wise Runs")
plt.ylabel("Runs")
plt.show()
```

```python
# Step 9: Top Wicket Takers
# Counted dismissals by bowler.
# Identified top 10 wicket-taking bowlers.
# Visualized using bar chart.
# Insight: Found leading bowlers in IPL history.
top_wicket = deliveries["bowler"].value_counts().head(10)

top_wicket.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Wicket Takers")
plt.ylabel("Wickets")
plt.show()
```
